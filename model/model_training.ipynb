```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, MobileNetV2
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# global variables
img_height, img_width = 224, 224 # do not change!
dataset_path='train_dataset'
bird_path=dataset_path+'/1'
nonbird_path=dataset_path+'/0'
```

## Printing out the resized training images

```python
# loads images and catches mac file system error
image_files = [f for f in os.listdir(bird_path) if not f.startswith('.DS_Store')]

# prints first 5 bird images
for i in range(5):
    img_path = os.path.join(bird_path, image_files[i])
    img = image.load_img(img_path, target_size=(224, 224))
    
    # Display the image using Matplotlib
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.axis('off')  # Turn off axis labels
    plt.title(image_files[i])  # Display image file name as title

plt.show()
```

```python
# same as previous cell but prints the non bird images
image_files = [f for f in os.listdir(nonbird_path) if not f.startswith('.DS_Store')]

for i in range(5):
    img_path = os.path.join(nonbird_path, image_files[i])
    img = image.load_img(img_path, target_size=(224, 224))
    
    # Display the image using Matplotlib
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.axis('off')  # Turn off axis labels
    plt.title(image_files[i])  # Display image file name as title

plt.show()
```

## Sets data split and image preprocessing method

```python
# preprocesses image pixels 
# divides pixel color value by 255 so that it is between 0-1
# makes the model run faster
def process(image):
    image=tf.cast(image/255. , tf.float32)
    return image

# sets the preprocessing function to the method just created
# also sets validation split to 20% (80% of images are for training, 20% are for validation)
datagen = ImageDataGenerator(
    preprocessing_function=process,
    validation_split=0.2
)
```

## Generates training and validation batches of data

```python
# training set
train_gen = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_height, img_width), # resize to 224x224
    batch_size=16, # number of training examples given per epoch
    class_mode='binary', # this is a binary classification problem
    subset='training', # defines that this is the training batch
    shuffle=True, # randomizes image order to remove some training bias
    seed=27 # randomization seed so that training is both random and reproducable
)

# same, but now for validation set
valid_gen = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_height, img_width),
    batch_size=16,
    class_mode='binary',
    subset='validation',
    shuffle=True,
    seed=27
)
```

## Importing the imagenet model and freezing its layers
We will be using the concept of transfer of learning in order to create an accurate computer vision model with a relatively tiny training size. To do this, we shall use the imagenet model, an extremely well trained model that can classify 1000 different objects accurately, to extract features. Then we will change the classification layers to be optimized for figuring out what picture is of a bird and what is not.

```python
base_model = MobileNetV2(weights='imagenet', # uses image nets weights
                         include_top=False, # removes last, classification layers
                         input_shape=(img_height, img_width, 3))

base_model.trainable = False # removes ability to retrain the imagenet model
```

```python
#base_model.summary()
```

## Creating the Model

```python
model = Sequential()

model.add(base_model) # adds imagenet base to our new model
model.add(Flatten()) # flatten into 1d layer
model.add(Dense(100,activation='relu')) # custom classification layers
model.add(Dense(1,activation='sigmoid')) # classification node

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

## Training the model and saving it's stats per epoch

```python
history = model.fit(
    train_gen,
    epochs=5,
    validation_data=valid_gen
)
```

## Training and Validation Accuracy

```python
plt.plot(history.history['accuracy'], color='blue', label='train')
plt.plot(history.history['val_accuracy'], color='orange', label='validation')
plt.legend()
plt.show()
```

## Testing unseen images

```python
dir = 'test_dataset'
image_files = [f for f in os.listdir(dir) if f.endswith('.jpg')]
image_files.sort()
all_preds = []

for image_name in image_files:
    image_path = os.path.join(dir, image_name)

    # Preprocess the image
    img = image.load_img(image_path, target_size=(img_height, img_width))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = img_array / 255.0  # Normalize pixel values

    # Make predictions
    predictions = model.predict(img_array)
    print(predictions[0])

    # Interpret predictions
    if predictions[0] >= 0.5:
        all_preds.append(1)
        print("Bird")
    else:
        all_preds.append(0)
        print("Not a bird")
```

```python
answers = [1,1,0,0,0,1,0]
print(answers)
print(all_preds)
accuracy = accuracy_score(answers, all_preds)*100

print(f"Accuracy: {accuracy:.2f}%")
```
